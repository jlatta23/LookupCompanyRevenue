```python
import pandas as pd
import pathlib
import numpy as np
```

```python
data_path = pathlib.Path("C:/WebDevelopment/businesslookup_research/data/")
```

```python
django_model_data = 'DjangoModelBusinessData_up_to_150_1.csv' #CHANGE THIS NAME FOR EACH

cols = ["BorrowerName", "BorrowerAddress",
        "BorrowerZip", "revenue_estimate",
        "industry_name", "code", 
        "JobsReported", "BorrowerState"]
df_business = pd.read_csv(data_path / django_model_data, usecols=cols)
```

```python
df_bus2 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_2.csv', usecols=cols)

df_bus3 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_3.csv', usecols=cols)

df_bus4 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_4.csv', usecols=cols)

df_bus5 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_5.csv', usecols=cols)

df_bus6 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_6.csv', usecols=cols)

df_bus7 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_7.csv', usecols=cols)

df_bus8 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_8.csv', usecols=cols)

df_bus9 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_9.csv', usecols=cols)

df_bus10 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_10.csv', usecols=cols)

df_bus11 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_11.csv', usecols=cols)

df_bus12 = pd.read_csv(data_path / 'DjangoModelBusinessData_up_to_150_12.csv', usecols=cols)

df_bus_big = pd.read_csv(data_path / 'DjangoModelBusinessDataBigCompanies.csv', usecols=cols)

df_business = df_business.append(df_bus2).append(df_bus3).append(df_bus4).append(df_bus5).append(df_bus6).append(df_bus7).append(df_bus8).append(df_bus9).append(df_bus10).append(df_bus11).append(df_bus12).append(df_bus_big)


```

```python
df_business['code'] = df_business['code'].fillna(0)
df_business['code'] = df_business['code'].astype(int)
df_business['code'] = df_business['code'].astype(str)
```

```python
df_business[df_business['BorrowerName'] == 'VULCAN MACHINERY CORPORATION']
```

```python
duplicate = df_business[df_business.duplicated(subset=['BorrowerName', 'BorrowerZip', 'industry_name'])]
```

```python
duplicate.head(50)
```

```python
len(duplicate)
```

```python
duplicate.to_csv(data_path / "duplicate_businesses.csv")
```

```python
df_business = df_business[df_business['BorrowerZip'].notna()]
df_business['BorrowerZip'] = df_business['BorrowerZip'].astype(str)
def clean_zip(zip_code):
    return zip_code.split('-')[0]
df_business['BorrowerZip'] = df_business['BorrowerZip'].apply(clean_zip)
```

```python
df_industry_avg = df_business.groupby("code")["revenue_estimate"].mean()
```

```python
df_industry_avg.head()
```

```python
df_industry_avg.to_csv(data_path / "avg_revenue_by_industry.csv")
```

```python
res = df_business.assign(percentile=df_business.groupby("code")['revenue_estimate'].rank(pct=True).mul(100))
```
