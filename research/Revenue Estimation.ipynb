```python
naics_approx_to_rev_payroll_ratio_df = pd.read_csv(data_path / 'naics_approximate_to_revenue_payroll_ratio.csv')
naics_approx_to_rev_payroll_ratio_df['naics_code_approximate'] = naics_approx_to_rev_payroll_ratio_df['naics_code_approximate'].astype(str)

def create_naics_col(row):
    naics = row['Industry'].split(' - ')[0].split(',')[0]
    return naics
df_productivity['NAICSCode'] = df_productivity.apply(create_naics_col, axis=1)

df_prod_pruned = df_productivity.drop(df_productivity[(df_productivity['Units'] != 'million$')].index)

seen_industry = {}
for index, row in df_prod_pruned.iterrows():
    if row['Industry'] in seen_industry:
        seen_industry[row['Industry']][row['Measure']] = row['2019']
    else:
        seen_industry[row['Industry']] = {row['Measure']: row['2019']}

naics_data_match = []
for k in seen_industry:
    naics_data_match.append((k.split(' - ')[0].split(',')[0]))
naics_data_match.append('45') #retail trade, I added this to productivity_by_industry.csv
```

```python
df_ppp['NAICSCode'] = df_ppp['NAICSCode'].astype(str)
def strip_str(row):
    return row['NAICSCode'].split('.')[0]
df_ppp['NAICSCode'] = df_ppp.apply(strip_str, axis=1)
```

```python
#Longest calculation, not too bad
l = df_ppp['NAICSCode']
naics_matching_code = []
jr=0
for naics in l:
    found_match = False
    for naics_match in naics_data_match:
        if naics == naics_match[:len(naics)]:
            naics_matching_code.append(naics_match)
            found_match = True
            break
        if naics_match[:3] == naics[:3]:
            naics_matching_code.append(naics_match)
            found_match = True
            break
        if naics_match[:4] == naics[:4]:
            naics_matching_code.append(naics_match)
            found_match = True
            break
        if naics_match[:5] == naics[:5]:
            naics_matching_code.append(naics_match)
            found_match = True
            break   
        if naics_match[:2] == naics[:2]:
            naics_matching_code.append(naics_match)
            found_match = True
            break
    if not found_match:
        naics_matching_code.append(np.nan)
        
    if jr %1000 == 0:
        print(jr)
    jr +=1
```

```python
df_ppp['Matching Code'] = naics_matching_code
```

```python
df_matched_ppp = pd.read_csv(data_path / naics_matched_ppp)
df_matched_ppp['Matching Code'] = df_matched_ppp['Matching Code'].fillna(0)
df_matched_ppp['Matching Code'] = df_matched_ppp['Matching Code'].astype(int)
df_matched_ppp['Matching Code'] = df_matched_ppp['Matching Code'].astype(str)
df_matched_ppp['NAICSCode'] = df_matched_ppp['NAICSCode'].fillna(0)
df_matched_ppp['NAICSCode'] = df_matched_ppp['NAICSCode'].astype(int)
df_matched_ppp['NAICSCode'] = df_matched_ppp['NAICSCode'].astype(str)

df_matched_ppp = pd.merge(df_matched_ppp, naics_approx_to_rev_payroll_ratio_df, how='left', left_on='Matching Code',
                         right_on='naics_code_approximate')
df_matched_ppp = pd.merge(df_matched_ppp, df_naics_codes, how='left', left_on='NAICSCode', right_on='code')
```

```python
def add_revenue_confidence(row):
    if  np.isnan(row['revenue_to_payroll']):
        return 'low'
    else:
        return 'high'
df_matched_ppp['revenue_confidence'] = df_matched_ppp.apply(add_revenue_confidence, axis=1)
```

```python
df_matched_ppp['payroll_annual'] = df_matched_ppp['CurrentApprovalAmount'] * 12.0 / 2.5
```

```python
df_matched_ppp['revenue_to_payroll'] = df_matched_ppp['revenue_to_payroll'].fillna(3.795) #fill in with average rev to payroll
df_matched_ppp['revenue_estimate'] = df_matched_ppp['payroll_annual'] * df_matched_ppp['revenue_to_payroll']
```

```python
def apply_unknown_industry(row):
    try:
        if np.isnan(row['name']):
            return 'Unknown'
    except Exception as e:
        return row['name']
df_matched_ppp['industry_name'] = df_matched_ppp.apply(apply_unknown_industry, axis=1)
```
